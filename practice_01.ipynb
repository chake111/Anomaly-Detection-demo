# Practice 01: IsolationForest 异常检测

本练习覆盖：导入库、生成二维数据、训练模型、预测异常并可视化。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
np.random.seed(42)

normal_data = np.random.randn(300, 2)
anomaly_data = np.array([
    [2.6, 2.8],
    [3.1, 2.5],
    [-2.9, -2.7],
    [2.8, -2.5],
    [-3.0, 2.7],
])

X = np.vstack([normal_data, anomaly_data])
y_true = np.array([0] * len(normal_data) + [1] * len(anomaly_data))

print('Data shape:', X.shape)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(normal_data[:, 0], normal_data[:, 1], s=20, label='Normal')
plt.scatter(anomaly_data[:, 0], anomaly_data[:, 1], s=80, c='red', label='Known Anomaly')
plt.title('Generated 2D Data')
plt.legend()
plt.show()
```

```python
model = IsolationForest(contamination=0.02, random_state=42)
model.fit(X)

pred = model.predict(X)
is_anomaly = pred == -1

print('Predicted anomaly count:', is_anomaly.sum())
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=is_anomaly, cmap='coolwarm', s=35)
plt.title('IsolationForest Prediction (True=Anomaly)')
plt.show()
```
